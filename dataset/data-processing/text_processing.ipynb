```python
import pandas as pd

pattern_pl = r'^(.*?) Autor: TEDx Talks ([\d\xa0 ,]*) wyświetle(?:ń|nia) (.*)$'
pattern_en = r'^(.*?) by TEDx Talks ([\d\xa0 ,]*) views (.*)$'
pattern_no_views = r'^(.*?) by TEDx Talks (.*)$'

data = []

for video in videos:
    text = video.replace('\n', ' ')
    if 'wyświetleń' in text or 'wyświetlenia' in text:
        match = re.match(pattern_pl, text)
    elif 'views' in text:
        match = re.match(pattern_en, text)
    else:
        match = re.match(pattern_no_views, text)

    if match:
        title = match.group(1)
        if 'views' in text or 'wyświetleń' in text or 'wyświetlenia' in text:
            views = match.group(2).replace('\xa0', '').replace(' ', '').replace(',', '')
            date = match.group(3)
        else:
            views = None
            date = match.group(2)
        data.append({'title': title, 'views': views, 'date_str': date})
    else:
        print('Nie udało się dopasować wyrażenia regularnego')
        print(video)

videos_df = pd.DataFrame(data)
```

```python
import re
from datetime import datetime
from dateutil.relativedelta import relativedelta

def parse_relative_date(date_str: str) -> datetime:
    base_date = datetime(2024, 12, 25)

    # Polskie i angielskie wzorce
    pattern_pl = re.compile(r'(\d+)\s+(rok|lata|lat|miesiąc|miesiące|miesięcy|tydzień|tygodnie|tygodni|dzień|dni|godzina|godziny|godzin|minut|minuta|minuty|sekund|sekundy|sekunda)\s+temu')
    pattern_en = re.compile(r'(\d+)\s+(year|years|month|months|week|weeks|day|days|hour|hours|minute|minutes|second|seconds)\s+ago')

    # Mapowanie polskich jednostek czasu
    units_pl = {
        'rok': 'years', 'lata': 'years', 'lat': 'years',
        'miesiąc': 'months', 'miesiące': 'months', 'miesięcy': 'months',
        'tydzień': 'weeks', 'tygodnie': 'weeks', 'tygodni': 'weeks',
        'dzień': 'days', 'dni': 'days',
        'godzina': 'hours', 'godziny': 'hours', 'godzin': 'hours',
        'minut': 'minutes', 'minuta': 'minutes', 'minuty': 'minutes',
        'sekund': 'seconds', 'sekundy': 'seconds', 'sekunda': 'seconds'
    }
    # Mapowanie angielskich jednostek czasu
    units_en = {
        'year': 'years', 'years': 'years',
        'month': 'months', 'months': 'months',
        'week': 'weeks', 'weeks': 'weeks',
        'day': 'days', 'days': 'days',
        'hour': 'hours', 'hours': 'hours',
        'minute': 'minutes', 'minutes': 'minutes',
        'second': 'seconds', 'seconds': 'seconds'
    }

    # Funkcja pomocnicza do przetwarzania dopasowań
    def process_match(match, units):
        value = int(match.group(1))
        unit = units[match.group(2)]
        return {unit: value * -1}

    # Przetwarzanie polskich wzorców
    matches_pl = pattern_pl.findall(date_str)
    matches_en = pattern_en.findall(date_str)

    kwargs = {}
    if matches_pl:
        for match in matches_pl:
            kwargs.update(process_match(re.match(pattern_pl, f"{match[0]} {match[1]} temu"), units_pl))
    elif matches_en:
        for match in matches_en:
            kwargs.update(process_match(re.match(pattern_en, f"{match[0]} {match[1]} ago"), units_en))
    else:
        # Brak dopasowania
        print('Nie udało się dopasować wzorca')
        print(date_str)
        return base_date

    return base_date + relativedelta(**kwargs)
```

```python
videos_df['date'] = videos_df['date_str'].apply(parse_relative_date)
```

```python
videos_df['year'] = videos_df['date'].dt.year
```
